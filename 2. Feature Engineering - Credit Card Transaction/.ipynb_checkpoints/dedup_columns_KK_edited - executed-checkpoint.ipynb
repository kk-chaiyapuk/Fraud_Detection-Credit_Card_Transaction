```python
import pandas as pd 
import numpy as np
import datetime as dt
import time
import glob
start_time = dt.datetime.now()
%matplotlib inline
```

```python
joined_files_2_ = glob.glob("output/2_*.csv")
joined_files_2_
```

```python
# I copied from here:
# https://stackoverflow.com/a/32961145
def duplicate_columns(frame):
    groups = frame.columns.to_series().groupby(frame.dtypes).groups
    dups = []
    for t, v in groups.items():
        dcols = frame[v].to_dict(orient="list")

        vs = list(dcols.values())
        ks = list(dcols.keys())
        lvs = len(vs)

        for i in range(1, lvs+1):
            for j in range(i+1, lvs+1):
                # KK: Check backward, because I want to keep the first column and drop later columns
                if vs[-i] == vs[-j]:
                    dups.append(ks[-i])
                    break

    return dups
```

```python
numvar = {}
listvar = []

def num_column_count(dataframe, name):
    
    global numvar, listvar
    
    numvars_new = len(dataframe.columns) - 2 # KK: The excluded 2 are record and fraud_label
    numvar[name] = numvars_new
    
    listvar_new = dataframe.columns.to_list()[2:] # KK: The excluded 2 are record and fraud_label
    listvar += listvar_new

    print("from Dataframe: " + name)
    print("new variables:", numvars_new,"  total variables:", sum(numvar.values()))
```

Please create "output_dedup" folder first

```python
%%time
begin=time.time()

number_of_file = len(joined_files_2_)

for num_for_dedup in range(1, number_of_file + 1):
    
    st=time.time()
    
    vars_prep_for_dedup = pd.read_csv(joined_files_2_[-num_for_dedup])
    new_filepath = joined_files_2_[-num_for_dedup].replace("output", "output_dedup").replace(".csv", "_dedup.csv")
    # This is 'output_dedup\\2_4_age_indicator_dedup.csv'
    
    for num_before in range(1 + num_for_dedup, number_of_file + 1):
        
        # Concat 2 files from backward
        # e.g. num_before: 'output\\2_3_maximum_indicator.csv'
        # with num_for_drop: 'output\\2_4_age_indicator.csv'
        df_concated = pd.concat([pd.read_csv(joined_files_2_[-num_before]).iloc[:,2:],
                                 vars_prep_for_dedup], axis = 1)
        
        dup_columns_name = duplicate_columns(df_concated)
        del df_concated
        
        vars_prep_for_dedup.drop(dup_columns_name, axis=1, inplace=True, errors='ignore')
        
    vars_prep_for_dedup.to_csv(new_filepath, index=False)
    
    new_filename = new_filepath.replace("_dedup.csv", "").split('\\')[1] # This is "2_4_age_indicator_dedup"
    num_column_count(vars_prep_for_dedup, new_filename)
    
    del vars_prep_for_dedup
    print(f'Run time for {new_filename}--------{time.time()-st:0.2f}s', end='\n\n')

print(f'Total run time:{(time.time()-begin)/60:0.2f}min')
```

```python
df_numvar = pd.DataFrame(numvar, index=["count_variables"]).T.iloc[::-1].reset_index()
df_numvar.loc['Total'] = df_numvar.sum(numeric_only=True)
df_numvar.columns = ['filename', 'count_variables']

df_numvar.to_csv('output_dedup/3_count_variables_dedup.csv')
df_numvar
```

```python
df_listvar = pd.DataFrame(listvar, columns = ["variables"]).drop_duplicates().iloc[::-1].reset_index(drop = True)

df_listvar.to_csv('output_dedup/3_list_variables_dedup.csv')
df_listvar
```
